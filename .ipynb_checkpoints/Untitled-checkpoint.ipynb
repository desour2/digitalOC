```python
import nflreadpy as nfl
import pandas as pd
import pyarrow
print(pyarrow.__version__)
everything = pd.DataFrame({})
for i in range(2020,2025):
    current = nfl.load_pbp(i)
    currentPanda = current.to_pandas()
    everything= pd.concat([currentPanda,everything], ignore_index = True)
print(everything.sort_values('game_date'))
```

```python
def getDataByHomeTeamName(data: pd.DataFrame, teamName: str):
    return data.loc[data['home_team'] == teamName].head(5)
```

```python
print(getDataByHomeTeamName(pbp_pd,"BUF"))
```

```python
def getDataByAwayTeamName(data: pd.DataFrame,teamName: str):
    return data.loc[data['away_team'] == teamName].head(100)
    
```

```python
print(getDataByAwayTeamName(pbp_pd,"ARI"))
```

```python
def areThesePlaysOutOfBounds(data: pd.DataFrame):
    for idx, row in data.iterrows():
        if row['out_of_bounds']:
            print('Play:',row)
```

```python
ariCard = getDataByAwayTeamName(pbp_pd,'ARI')
areThesePlaysOutOfBounds(ariCard)
```

```python
def printColumn(data: pd.DataFrame, column: str):
    print(data[column])
```
