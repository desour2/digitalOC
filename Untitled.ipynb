```python
def filter_true_plays(pbp: pd.DataFrame) -> pd.DataFrame:
    """Keep real offensive plays (run/pass), drop penalties and weird stuff."""
    df: pd.DataFrame = pbp.copy()
    if "pass" not in df.columns:
        df["pass"] = (df["play_type"] == "pass").astype(int)
    if "rush" not in df.columns:
        df["rush"] = (df["play_type"] == "run").astype(int)

    df = df[
        ((df["pass"] == 1) | (df["rush"] == 1)) &
        (df["penalty"] == 0)
    ].copy()

    # Sorting guarantees time order for rolling/expanding features.
    df.sort_values(["season", "week", "game_id", "play_id"], inplace=True, kind="mergesort")
    return df
```
